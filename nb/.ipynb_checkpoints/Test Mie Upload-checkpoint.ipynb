```python
from IPython.display import Image
import ee
import ee.mapclient

ee.Initialize()

collection = ee.ImageCollection("users/fabiolananotizie/test_upload/collezione")
collection.getInfo()
```

```python
#Convert the collection to a list and get the number of images.
size = collection.toList(100).length()
'Number of images: ', size
```

```python
#Get the number of images.
count = collection.size()
type(count)
```

```python
#Create a collection by giving a list to the constructor.
b2 = ee.Image("users/fabiolananotizie/test_upload/collezione/LC80430332016252LGN00_B2")
b3 = ee.Image("users/fabiolananotizie/test_upload/collezione/LC80430332016252LGN00_B3")
b4 = ee.Image("users/fabiolananotizie/test_upload/collezione/LC80430332016252LGN00_B4")
b5 = ee.Image("users/fabiolananotizie/test_upload/collezione/LC80430332016252LGN00_B5")
b6 = ee.Image("users/fabiolananotizie/test_upload/collezione/LC80430332016252LGN00_B6")
```

```python
region = [[-25.0, -37.0], [60.0, -41.0], [58.0, 39.0], [-31.0, 38.0], [-25.0, -37.0]]
collectionFromConstructor = ee.ImageCollection([b2, b3, b4, b5, b6]);
median = collectionFromConstructor.median();
url_median = median.getThumbUrl({'region':region})
Image(url=url_median)
```

```python
immagine = ee.Image([b5,b4,b3]);
url_tc = immagine.getThumbUrl({
    'bands' : 'b1,b1_1,b1_2',
    'min' :1000,
    'max':20000})
Image(url=url_tc)
```

```python
vizParam = {'palette':'a1661e,b1971b,e48e58,f7cf71,44e448,7fc477,33ad37,438c8f,1ca127',
            'min' :-.25,
            'max':1,
           };

ndvi = immagine.normalizedDifference(['b1', 'b1_2']);
url_ndvi = ndvi.getThumbUrl(vizParam);
Image(url=url_ndvi)
```
